```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def NR_root(func,x_0,epsilon=1.0e-10,nmax=1000):
    
    x_n_m_1=x_0
    
    # Loop until maximum so we don't go on forever
    for n in range(nmax):
            
        try:
            x_n = x_n_m_1 - func(x_n_m_1)/dfunc(x_n_m_1)       
        
        # Make sure we don't divide by zero
        except ZeroDivisionError:
            print('Divide by zero, try a different guess!')
    
        print('New guess at step: '+str(n)+' is: ' + str(x_n))
        
        # Check for convergence
        if np.abs(x_n) < epsilon and np.abs(x_n-x_n_m_1) < epsilon:
            print('Root found: '+str(x_n))
            return x_n
        elif np.abs(x_n-x_n_m_1) < np.abs(epsilon*x_n):
            print('Root found: '+str(x_n))
            return x_n
        else:
            x_n_m_1 = x_n
    
    # Give an error if we 
    raise Exception('After '+str(nmax)+' steps, the root could not converge below '+str(epsilon))
    
```

```python
# Define our function and derivative
def func(x):
    return x**3+6

def dfunc(x):
    return 3*x**2

# Check out the function

x=np.linspace(-20,20,100)

fig1,ax1=plt.subplots(figsize=(5,5))
ax1.plot(x,func(x))

plt.show()
```

```python
# Choose an initial guess and find the root
x_0=5
x_n=NR_root(func,x_0)
```

```python
# Lets test it! Plug back in.
print(func(x_n))
```
