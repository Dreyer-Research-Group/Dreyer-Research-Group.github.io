```python
# Imports
import numpy as np
import matplotlib.pyplot as plt

plt.rc('text', usetex=True)
plt.rc('text.latex', preamble=r'\usepackage{cmbright}')
plt.rc('font',**{'family':'sans-serif','sans-serif':['Arial'],'size':18})
```

```python
def linear_congruent(seed,a,b,c,nNumbers):
    '''Generates nNumbers (should be even) pseudorandom numbers from seed'''
    
    rands=np.zeros(nNumbers)
    rands[0]=seed
    
    for r in range(1,nNumbers):
        rands[r]=(a*rands[r-1]+b) % c
    
    return rands/c
```

```python
a=7.**5
b=0.
c=2**31-1
nNumbers=1000 # Should be even, since we split into two below

seed=0.5
rands=linear_congruent(seed,a,b,c,nNumbers)

seed=0.51
rands2=linear_congruent(seed,a,b,c,nNumbers)


fig1,(ax1,ax2)=plt.subplots(2,1,figsize=(5,10))
ax1.plot(rands,'o',label='Seed = 0.5')
ax1.plot(rands2,'o',label='Seed = 56')

rands=np.reshape(rands,[nNumbers//2,2])
rands2=np.reshape(rands2,[nNumbers//2,2])

ax2.plot(rands[:,0],rands[:,1],'o',label='Seed = 0.5')
ax2.plot(rands2[:,0],rands2[:,1],'o',label='Seed = 56')

ax1.set_xlabel('$i$')
ax1.set_ylabel('$x_i$')
ax2.set_xlabel('$x_i$')
ax2.set_ylabel('$x_{i+1}$')
ax1.legend()

#plt.savefig('lin_cong_rand.pdf',bbox_inches='tight')
plt.show()
```
